```
import matplotlib.pyplot as plt

plt.plot(d[:,1])
plt.show
```

```
#MA filter with W = 5

series = d[:,1]
new_series = []
for i in range(0,series.shape[0]):
  if i>5:
    new_series.append((series[i]+series[i-1]+series[i-2]+series[i-3]+series[i-4])/5)
ns = np.array(new_series, dtype='float32')
plt.plot(range(series.shape[0]),series)
plt.plot(range(ns.shape[0]),ns)
plt.show()
```
