```
import matplotlib.pyplot as plt
import numpy as np

eeg_data = data['eeg_data']
labels = data['labels']
```

```
print(eeg_data.shape)
print(eeg_data[396].shape)
print(eeg_data[396][0].shape)
plt.plot(eeg_data[396][0])
```

```
sample = 175
column = 0

temp_col = []
window = 10
for i in range(eeg_data[sample][column].shape[0]-window+1):
  avg = sum(eeg_data[sample][column][i:i+window]) / window
  temp_col.append(avg)
  
  
  
d_col = eeg_data[sample][column][::5]
temp_col_ = []
window = 10
for i in range(d_col.shape[0]-window+1):
  avg = sum(d_col[i:i+window]) / window
  temp_col_.append(avg)
```

```
#Downsample then MAF
plt.plot(temp_col_, color = 'r')
print(np.asarray(temp_col_).shape)

#MAF then Downsample
plt.plot(temp_col[::5], color = 'b')
print(np.asarray(temp_col[::5]).shape)

#only downsampled signal
plt.plot(d_col.reset_index()[0],color='g')
```
